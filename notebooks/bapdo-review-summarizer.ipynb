```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

# API 키 설정
if not os.getenv("GOOGLE_API_KEY"):
    raise RuntimeError("GOOGLE_API_KEY 환경변수를 설정한 뒤 실행하세요.")

# 프롬프트 템플릿 설정
prompt = PromptTemplate(
    input_variables=["question"],
    template="질문에 답하세요:\n{question}"
)

# LangChain용 Gemini 모델 설정
llm = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0.7)

# 체인 구성
chain = LLMChain(llm=llm, prompt=prompt)

# 실행 예시
response = chain.run("부산에서 가볼 만한 문화 공간은?")
print(response)
```

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

# API 키 설정
if not os.getenv("GOOGLE_API_KEY"):
    raise RuntimeError("GOOGLE_API_KEY 환경변수를 설정한 뒤 실행하세요.")

# 프롬프트 템플릿 구성
prompt = PromptTemplate(
    input_variables=["question"],
    template="다음 질문에 한국어로 정중하게 답하세요:\n\n{question}"
)

# Gemini-1.5-Flash 모델 연결
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",  # 여기서 모델만 바꿔주면 됨
    temperature=0.7
)

# 체인 구성
chain = LLMChain(llm=llm, prompt=prompt)

# 테스트 실행
response = chain.run("부산에서 가볼 만한 전시회나 공연을 알려줘.")
print(response)
```
